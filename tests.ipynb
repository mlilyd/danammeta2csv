```python
spacy.prefer_gpu()
nlp = spacy.load("en_core_web_trf")
```

```python
def nlp_segments(caption):
    doc = nlp(caption, )
    info = {}
    info['caption'] = caption
    info['noun_phrases'] = list(doc.noun_chunks)

    entities = []
    for entity in doc.ents:
        entities.append((entity.text, entity.label_))

    info['entities'] = entities

    return info
```
